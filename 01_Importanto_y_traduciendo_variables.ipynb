**Traduciendo variables**


```python
datos.columns
```

```python
dic_columnas = {
    'LIMIT_BAL' : 'limite',
    'CHECKING_ACCOUNT' : 'cuenta_corriente',
    'EDUCATION' : 'escolaridad',
    'MARRIAGE' : 'estado_civil',
    'AGE' : 'edad',
    'BILL_AMT' : 'valor_factura',
    'PAY_AMT' : 'valor_pago',
    'DEFAULT' : 'moroso'
}
```

```python
tarjetas = datos.rename(columns = dic_columnas)
tarjetas.head()
```

cambio de nombre de las variables de los valores

```python
tarjetas.cuenta_corriente.unique()
```

```python
dic_cuenta = {
    'Yes' : 'si', 'No': 'no'
}
```

```python
#Usamos map para cambiar los datos de las variables
#pero al no atribuir a la base de datos solamente genera los nuevos valores.
tarjetas.cuenta_corriente.map(
    dic_cuenta
)
```

```python
tarjetas.head()
```

```python
tarjetas.cuenta_corriente = tarjetas.cuenta_corriente.map(
    dic_cuenta
)
```

```python
tarjetas.head()
```

```python
tarjetas.escolaridad.unique()
```

```python
dic_escolaridad = {
    '2.University' : '2.Universidad',
    '3.Graduate School' : '3.Post-grado',
    '1.High School' : '1.Colegio'
}
```

```python
tarjetas.escolaridad = tarjetas.escolaridad.map(
    dic_escolaridad
)
tarjetas.head()
```

```python
tarjetas.estado_civil.unique()
```

```python
tarjetas.estado_civil = tarjetas.estado_civil.map(
    {
       'Married': 'casado', 'Single' : 'soltero'
    }
)
```

```python
tarjetas.head()
```

```python
tarjetas.edad.unique()
#Al tener tantos datos, no es factible utilizar los graficos visto anteriormente
#Como solucion se unir√° por rango de edad
```

```python
tarjetas.edad.unique().max()
```

```python
tarjetas.edad.unique().min()
```

```python
bins=[20,30,40,50,80]
nombres = ['20-30' ,'30-40' ,'40-50' ,'50+']

tarjetas['rango_edad'] = pd.cut(
    tarjetas.edad,
    bins,
    labels = nombres
    )

tarjetas.head()
```

```python
sns.boxplot(
    data = tarjetas,
    x= 'rango_edad',
    y= 'limite'
)
```
